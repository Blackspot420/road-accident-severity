```python
import pandas as pd

# Load the CSV file
df = pd.read_csv('accident_data.csv')

# Display the first few rows
df.head()
```

```python
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import LabelEncoder
import joblib


label_cols = ['Weather_Condition', 'Road_Surface_Condition', 'Light_Condition']
df_encoded = df.copy()
for col in label_cols:
    df_encoded[col] = LabelEncoder().fit_transform(df[col])


X = df_encoded[['Weather_Condition', 'Road_Surface_Condition', 'Vehicle_Speed', 'Light_Condition', 'Number_of_Vehicles', 'Driver_Age']]
y = df_encoded['Accident_Severity']


X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=0)


model = LinearRegression()
model.fit(X_train, y_train)

# Save model
joblib.dump(model, 'accident_severity_model.pkl')

print("✅ Model trained and saved.")
```

```python
import pandas as pd
import joblib

# Sample input
sample_input = pd.DataFrame([{
    'Weather_Condition': 1,       # Example: Rainy
    'Road_Surface_Condition': 1,  # Wet
    'Vehicle_Speed': 65,
    'Light_Condition': 1,         # Darkness
    'Number_of_Vehicles': 2,
    'Driver_Age': 30
}])

# Load the saved model
model = joblib.load('accident_severity_model.pkl')

# Make a prediction
prediction = model.predict(sample_input)

print(f"🔮 Predicted Accident Severity: {prediction[0]:.2f}")
```
